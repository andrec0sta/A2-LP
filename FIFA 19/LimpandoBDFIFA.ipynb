```python
def f(x):
    unit = x[len(x)-1:]
    x = x[1:len(x)-1]
    if unit == 'M':
        return float(x)*1000
    elif unit == 'K':
        return float(x)*1

lista = []
for a in df["Value"]:
    lista.append(f(a))
    
df['Value'] = lista    

            
print(df['Value'])
```
