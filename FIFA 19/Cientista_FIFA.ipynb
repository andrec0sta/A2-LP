## Modelo Estatisco

### 1.  É familar para um telespectador assíduo de jogos de futebol que o camisa 1 geralmente é o goleiro. Então, segundo os dados, qual seria posição em que os camisas 1 jogam?
Para responder, primeiro vamos separar os camisas 1 de todos os outros numeros,e em seguida agrupá-los por posição.

```python
a = df[['PosicionamentoCampo', 'Jersey_Number']]
a = a[a['Jersey_Number'] == 1 ]
q1 = a.groupby('PosicionamentoCampo').count()
q1.head()
```

Pelo resultado, vemos que todos os jogadores que utilizam camisa de numero 1 são goleiros.

### 2. É de consenso geral que frequentemente o melhor jogador do time é o camisa 10. Segundo os dados, qual é a posição mais  frequente do camisa 10?

Para responder a essa pergunta, teremos que fazer um processo similar ao feito na questão 1

```python
a = df[['PosicionamentoCampo', 'Jersey_Number']]
a = a[a['Jersey_Number'] == 10 ]
q2 = a.groupby('PosicionamentoCampo').count()
q2.head()
```

Pelo resultado acima, observamos que o a região que mais tem camisas 10 é no meio.

Pode-se concluir que o meio é lugar mais valorizado do futebol.

### 3. Qual é o número que tem melhor avaliação? E o maior preço de contratação?

Para ínicio, vamos pegar a tabela de dados e agrupá-la por preço, por número da camisa e fazer a média da avaliação e do preço de cada jogador.

```python
a = df[['Jersey_Number','Overall','Value']]
q3 = a.groupby('Jersey_Number').mean()
q3.head(12)
```

```python
print(q3.max())
```

```python
print(q3[q3['Overall']>70])
```

```python
print(q3[q3['Value']>5000])
```

Portanto o número em que a média dos jogadores possui a melhor avaliação é o 79, entretanto esse número é pouco utilizado o futebol, deixando ainda uma grande importancia para a camisa 10, que é uma das mais usadas.

Já sobre os valores, os camisas 10 realmente levaram a vantagem.

### 4. Existe alguma posição que os jogadores são mais valorizados?

Pensaremos parecido com a anterior, mas em vez do número que o jogador usa, vamos analisar a região em que ele joga.

```python
a = df[['PosicionamentoCampo','Overall','Value']]
q4 = a.groupby('PosicionamentoCampo').mean()
q4.head()
```

Pela tabela observamos que a avaliação do jogador é bem menor quando ele joga no gol. Além disso, vemos que por preço, os jogadores de meio de campo e de ataque são bem mais valorizados que os zagueiros(defensores) que, por sua vez, valem bem mais que os goleiros.

### 5. Em qual idade os jogadores são mais valorizados?

```python
a = df[['Age','Overall','Value']]
q5 = a.groupby('Age').mean()
q5.head(30)
```

Vemos que os jogadores quando estão nos extremos da carreira valem menos. E chegam ao topo por volta dos 24 aos 31 anos. 
A avaliação tambem segue mais ou menos mesmo criterio

Vamos observar no caso especifico de goleiros, que geralmente tem as carreiras mais longas

```python
a = df[['Age','Position','Overall','Value']]
a = a[a['Position']=='GK']
q5g = a.groupby('Age').mean()
q5g.head(30)
```

No gol, observamos que o jogador demora mais tempo para atingir o ápice da carreira que geralmente acontece aos 28 anos e vai até por volta dos 35 anos. Tendo exceções que podem continuar em alto nivel por mais tempo, exemplo disso é que goleiros com idade de 39 anos têm um "overall" tão alto quanto outros jogadores no ápice. Com idade de 45, deve haver uma unica exceção por isso tão altos niveis de avaliação e de valor.

### 6. Tem alguma forma de prever o preço de um jogador?

Como vimos nas questões acima, diferentes regiões têm diferentes valores para diferentes paramêtros. Então, para tentar prever os preços dos jogadores, vamos dividir em diferentes regiões que eles jogam. 

```python
novo = df[[ 'Age', 'Overall','Value','Weak_Foot', 'Skill_Moves','Crossing', 'Finishing', 'HeadingAccuracy', 'ShortPassing', 'Volleys',
       'Dribbling', 'Curve', 'FKAccuracy', 'LongPassing', 'BallControl','Acceleration', 'SprintSpeed', 'Agility', 'Reactions', 'Balance',
       'ShotPower', 'Jumping', 'Stamina', 'Strength', 'LongShots','Aggression', 'Interceptions', 'Positioning', 'Vision', 'Penalties',
       'Composure', 'Marking', 'StandingTackle', 'SlidingTackle', 'GKDiving','GKHandling', 'GKKicking', 'GKPositioning', 'GKReflexes',
       'PosicionamentoCampo']]

novo = novo.dropna(0)
```

```python
from sklearn import linear_model,model_selection
```

#### TOTAL

```python
x = novo.drop(['Value','PosicionamentoCampo'],axis=1)
y = novo['Value']

x_treinot,x_testet,y_treinot,y_testet = model_selection.train_test_split(x,y,test_size=.4,random_state=1)
```

```python
modelt = linear_model.LinearRegression()
modelt.fit(x_treinot,y_treinot)
```

```python
print(modelt.score(x_treinot,y_treinot))
```

#### GOLEIROS

```python
gol = novo[novo['PosicionamentoCampo']=='GOL']
x = gol[['GKDiving','GKHandling', 'GKKicking', 'GKPositioning', 'GKReflexes','Overall','Age']]
y = gol['Value']

x_treinog,x_testeg,y_treinog,y_testeg = model_selection.train_test_split(x,y,test_size=.4,random_state=1)

modelg = linear_model.LinearRegression()
modelg.fit(x_treinog,y_treinog)

print(modelg.score(x_treinog,y_treinog))
```

#### DEFENSORES

```python
defe = novo[novo['PosicionamentoCampo']=='DEF']
x = defe[['Age', 'Overall','Weak_Foot', 'Skill_Moves','Crossing', 'Finishing', 'HeadingAccuracy', 'ShortPassing', 'Volleys',
       'Dribbling', 'Curve', 'FKAccuracy', 'LongPassing', 'BallControl','Acceleration', 'SprintSpeed', 'Agility', 'Reactions', 'Balance',
       'ShotPower', 'Jumping', 'Stamina', 'Strength', 'LongShots','Aggression', 'Interceptions', 'Positioning', 'Vision', 'Penalties',
       'Composure', 'Marking', 'StandingTackle', 'SlidingTackle']]
y = defe['Value']

x_treinod,x_tested,y_treinod,y_tested = model_selection.train_test_split(x,y,test_size=.4,random_state=1)

modeld = linear_model.LinearRegression()
modeld.fit(x_treinod,y_treinod)

print(modeld.score(x_treinod,y_treinod))
```

#### MEIAS

```python
mec = novo[novo['PosicionamentoCampo']=='MEC']
x = mec[['Age', 'Overall' ,'Weak_Foot', 'Skill_Moves','Crossing', 'Finishing', 'HeadingAccuracy', 'ShortPassing', 'Volleys',
       'Dribbling', 'Curve', 'FKAccuracy', 'LongPassing', 'BallControl','Acceleration', 'SprintSpeed', 'Agility', 'Reactions', 'Balance',
       'ShotPower', 'Jumping', 'Stamina', 'Strength', 'LongShots','Aggression', 'Interceptions', 'Positioning', 'Vision', 'Penalties',
       'Composure', 'Marking', 'StandingTackle', 'SlidingTackle']]
y = mec['Value']

x_treinom,x_testem,y_treinom,y_testem = model_selection.train_test_split(x,y,test_size=.4,random_state=1)

modelm = linear_model.LinearRegression()
modelm.fit(x_treinom,y_treinom)

print(modelm.score(x_treinom,y_treinom))
```

#### ATACANTES

```python
atq = novo[novo['PosicionamentoCampo']=='ATQ']
x = atq[['Age',  'Overall','Weak_Foot', 'Skill_Moves','Crossing', 'Finishing', 'HeadingAccuracy', 'ShortPassing', 'Volleys',
       'Dribbling', 'Curve', 'FKAccuracy', 'LongPassing', 'BallControl','Acceleration', 'SprintSpeed', 'Agility', 'Reactions', 'Balance',
       'ShotPower', 'Jumping', 'Stamina', 'Strength', 'LongShots','Aggression', 'Interceptions', 'Positioning', 'Vision', 'Penalties',
       'Composure', 'Marking', 'StandingTackle', 'SlidingTackle']]
y = atq['Value']

x_treinoa,x_testea,y_treinoa,y_testea = model_selection.train_test_split(x,y,test_size=.4,random_state=1)

modela = linear_model.LinearRegression()
modela.fit(x_treinoa,y_treinoa)

print(modela.score(x_treinoa,y_treinoa))
```
